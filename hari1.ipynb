```python
import pandas as pd 

#load dataset csv
df = pd.read_csv("iris.csv")

#tampilkan 10 baris pertama 
print(df.head(10))
```

```python
import matplotlib.pyplot as plt

# Scatter Plot
plt.scatter(df["sepal_length"], df["sepal_width"], c="blue", alpha=0.5)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# Line Plot
plt.plot(df["sepal_length"].head(20), label="Sepal Length", marker="o")
plt.plot(df["sepal_width"].head(20), label="Sepal Width", marker="s")
plt.title("Line Plot: Sepal Length & Sepal Width (first 20 rows)")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()

```
